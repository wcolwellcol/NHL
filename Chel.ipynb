```python
def shtpct(df, name):
    df_home = df.loc[df.HOME == name]
    df_away = df.loc[df.AWAY == name]
    total_goals = df_away['AWAY SCORE'].sum() + df_home['HOME SCORE'].sum()
    total_shots = df_away['AWAY SHOTS'].sum() + df_home['HOME SHOTS'].sum()
    return total_goals/total_shots
```

```python
print(shtpct(df, 'THEO'))
print(shtpct(df, 'WILL'))
print(shtpct(df, 'MARK'))
```

```python
def shtpctallowed(df, name):
    df_home = df.loc[df.HOME == name]
    df_away = df.loc[df.AWAY == name]
    total_goals = df_away['HOME SCORE'].sum() + df_home['AWAY SCORE'].sum()
    total_shots = df_away['HOME SHOTS'].sum() + df_home['AWAY SHOTS'].sum()
    return total_goals/total_shots
```

```python
print(shtpctallowed(df, 'THEO'))
print(shtpctallowed(df, 'WILL'))
print(shtpctallowed(df, 'MARK'))
```

```python
def games_played(name):
    return df.loc[(df.HOME==name)|(df.AWAY==name)]
```

```python
games_played('THEO')
```

```python
def wins(name):
    df_name_wins = df.loc[(df.HOME == name) & (df['HOME SCORE'] > df['AWAY SCORE'])]
    df_name_wins_2 = df.loc[(df.AWAY == name) & (df['AWAY SCORE'] > df['HOME SCORE'])]
    df_name_wins_res = pd.concat([df_name_wins, df_name_wins_2])
    return df_name_wins_res
```

```python
wins('WILL')
```

```python
def losses(name):
    df_name_losses = df.loc[(df.HOME == name) & (df['HOME SCORE'] < df['AWAY SCORE'])]
    df_name_losses_2 = df.loc[(df.AWAY == name) & (df['AWAY SCORE'] < df['HOME SCORE'])]
    df_name_losses_res = pd.concat([df_name_losses, df_name_losses_2])
    return df_name_losses_res
```

```python
losses('MARK')
```

```python
def winning_pct(name):
    return wins(name).shape[0]/games_played(name).shape[0]
```

```python
print(winning_pct('THEO'))
print(winning_pct('WILL'))
print(winning_pct('MARK'))
```

```python
def avg_goals_scored(name):
  return games_played(name).apply(lambda x: x['HOME SCORE'] if x['HOME']==name else x['AWAY SCORE'],axis=1).values.mean()

def avg_goals_allowed(name):
  return games_played(name).apply(lambda x: x['HOME SCORE'] if x['AWAY']==name else x['HOME SCORE'],axis=1).values.mean()
```

```python
print(avg_goals_scored('THEO'))
print(avg_goals_scored('WILL'))
print(avg_goals_scored('MARK'))
print()
print(avg_goals_allowed('THEO'))
print(avg_goals_allowed('WILL'))
print(avg_goals_allowed('MARK'))
```
